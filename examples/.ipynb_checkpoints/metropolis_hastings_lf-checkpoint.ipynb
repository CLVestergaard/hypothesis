# Likelihood-free Metropolis-Hastings Posterior Inference

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

from torch.distributions.normal import Normal
```

## Observations

```python
num_observations = 100
theta_true = 1.


N = torch.distributions.normal.Normal(theta_true, 1.)
observations = N.sample(torch.Size([num_observations])).view(-1, 1)

plt.hist(observations.numpy(), bins=50, density=True)
plt.show()
```

## Likelihood-free likelihood-ratio (no pretraining)

```python
def allocate_classifier(hidden=100):
    return torch.nn.Sequential(
        torch.nn.Linear(1, hidden),
        torch.nn.ReLU(),
        torch.nn.Linear(hidden, hidden),
        torch.nn.ReLU(),
        torch.nn.Linear(hidden, hidden),
        torch.nn.ReLU(),
        torch.nn.Linear(hidden, hidden),
        torch.nn.ReLU(),
        torch.nn.Linear(hidden, 1),
        torch.nn.Sigmoid())
```

```python
def train_classifier(theta_next, theta, gamma=1.):
    batch_size = 32
    num_samples = 10000
    epochs = 5
    iterations = int(num_samples * epochs / batch_size)
    real = torch.ones(batch_size, 1)
    fake = torch.zeros(batch_size, 1)
    bce = torch.nn.BCELoss()
    classifier = allocate_classifier()
    optimizer = torch.optim.Adam(classifier.parameters())
    size = torch.Size([batch_size])
    N_next = Normal(theta_next, 1.)
    N = Normal(theta, 1.)
    for iteration in range(iterations):
        x_real = N_next.sample(size).view(-1, 1)
        x_fake = N.sample(size).view(-1, 1)
        y_real = classifier(x_real)
        y_fake = classifier(x_fake)
        loss = (bce(y_real, real) + bce(y_fake, fake)) / 2.
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    return classifier
```

```python
def likelihood_ratio(observations, theta_next, theta):
    classifier = train_classifier(theta_next, theta)
    s = classifier(observations)
    lr = ((s) / (1 - s)).log().sum().exp()
    N_next = Normal(theta_next, 1.)
    N = Normal(theta, 1.)
    analytical = (N_next.log_prob(observations).sum() - N.log_prob(observations).sum()).exp()
    print(theta, "->", theta_next, ':', lr.item(), " -- analytical:", analytical.item())
    return lr
```

```python
from hypothesis.transition import NormalTransitionDistribution
from hypothesis.inference import RatioMetropolisHastings

transition = NormalTransitionDistribution(.1)
sampler = RatioMetropolisHastings(likelihood_ratio, transition)
```

```python
theta_0 = torch.tensor(5).float()
result = sampler.infer(
    observations,
    theta_0=theta_0,
    samples=5000,
    burnin_steps=1000)
```

```python
from hypothesis.visualization.mcmc import plot_trace
from hypothesis.visualization.mcmc import plot_density
from hypothesis.visualization.mcmc import plot_autocorrelation

plot_trace(result, show_mean=True, truth=theta_true, show_burnin=True, aspect=1)
plt.show()
plot_density(result, show_mean=True, truth=theta_true, bins=40)
plt.show()
plot_autocorrelation(result, max_lag=100, interval=1)
plt.show()
print(result.effective_size())
print(result.efficiency())
thinned = result.thin()
plot_autocorrelation(thinned, max_lag=100, interval=1)
plt.show()
plot_density(thinned, show_mean=True, truth=theta_true, bins=40)
plt.show()
print(thinned.effective_size())
print(thinned.efficiency())
```
